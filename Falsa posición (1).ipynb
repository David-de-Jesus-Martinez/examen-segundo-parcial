```python
def f(x):
  return (2*x**3) - (11.7*x**2) + (17.7*x) - 5
```

```python
def falsa(xl, xu, criterio_paro = 0.001):
    xl = xl
    xu = xu
   
    f_xu = f(xu)
    
    f_xl = f(xl)
    
    error = xu
    xr_ant = xu
    raiz = None
    if f_xl * f_xu < 0:
        while(error > criterio_paro):
            xr = (xu - ((f_xu * (xl - xu) / (f_xl - f_xu))))
            xrVec.append(xr)
            f_xr = f(xr)
            if f_xl * f_xr < 0:
                xu = xr
                f_xu = f_xr
            if f_xu * f_xr < 0:
                xl = xr
                f_xl = f_xr
            error = abs((xr-xr_ant)/xr)
            raiz = xr
            xr_ant = xr
    else:
        print("La raíz no está en el intervalo")
    return raiz
```
