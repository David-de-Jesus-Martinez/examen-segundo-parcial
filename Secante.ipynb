```python
import matplotlib.pyplot as plt
import math
```

```python
def f(xi):
    return (2*xi**3) - (11.7*xi**2) + (17.7*xi) - 5

def xi_next(xi0,xi1,Criterio_Paro = 0.0001):
    x = [xi0, xi1]
    Error=[]
    No_Converge=0
    Converge=True
    mensaje='Sí Converge'
    i=1
   
    Error.append(abs(x[i]-x[i-1]))
    while((Error[i-1]>Criterio_Paro) or No_Converge==50 or x[i]>999999999999):
        x.append(x[i] - (f(x[i])*(x[i-1]-x[i]))/(f(x[i-1])-f(x[i])) )
        i+=1
        Error.append(abs(x[i]-x[i-1]))
        if(No_Converge==50 or x[i]>999999999999):
                mensaje = 'No converge'
                Converge=False
        No_Converge+=1
    
    print("Raíz: "+ str(x) +"\n\nError: "+ str(Error))
    plt.plot(x,label="x Raíz",color='b')
    plt.plot(Error,label="Error",color='r')
    plt.title("Secante " + mensaje, fontsize = 24)
    plt.legend()
    plt.show()
    
    return Converge
    
x0 = int(input('Introduce el valor de x0 \n'))
x1 = int(input('Introduce el valor de x1 \n'))
s=int(input("Selecciona 1 si quieres introducir un criterio de paro de lo contrario escribe cualquier numero \n"))
if s==1:
    criterio_paro = float(input('Introduce el valor del criterio de paro \n'))
    print(str(xi_next(x0,x1,criterio_paro)))
else:    
   print(str(xi_next(x0,x1)))
```
