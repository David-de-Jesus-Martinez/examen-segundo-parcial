```python
import matplotlib.pyplot as plt
import math

def g1(xn, criterio_paro = 0.0001):
    No_Converge=0
    converge = True
    mensaje = 'Converge'
    x = [xn]
    Error=[]
    n=0
    x.append(((2*x[n]**3) - (11.7*x[n]**2) - 5) / - 17.7)
    n+=1
    Error.append(abs(x[n] - x[n-1]))
    
    while ((Error[n-1] > criterio_paro) and No_Converge<50 and x[n]<999999999999):
            
            x.append(((2*x[n]**3) - (11.7*x[n]**2) - 5) / - 17.7)
            n+=1
            No_Converge+=1
            Error.append(abs(x[n] - x[n-1]))
            if(No_Converge==50 or x[n]>999999999999):
                mensaje = 'No converge'
                converge = False            
            
    print('g')
    print('x: '+ str(x) )
    print('\n Error: ' + str(Error))
   
    plt.plot(x,color='b')
    plt.plot(Error,color='r')
    plt.title("g " + mensaje, fontsize = 24)
    plt.xlabel("Rojo = Vector del Error", fontsize = 14)
    plt.ylabel("Azul = Raiz", fontsize = 14)
    plt.show()
    
    return converge

xn = int(input('Introduce el valor incial de x \n'))
s=int(input("Selecciona 1 si quieres introducir un criterio de paro de lo contrario escribe cualquier numero \n"))
if s==1:
    criterio_paro = float(input('Introduce el valor del criterio de paro \n'))
    print(str(g1(xn,criterio_paro)))
else:    
    print(str(g1(xn)))
```
